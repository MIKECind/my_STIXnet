```python
import spacy
from spacy import displacy

# 加载英语语言模型
nlp = spacy.load('en_core_web_sm')

# 定义要构建依赖关系图的文本
text = "FIN7 malware has created Registry Run and RunOnce keys to establish persistence, and has also added items to the Startup folder."

# 处理文本
doc = nlp(text)

# 遍历每个句子并构建依赖关系图
for sent in doc.sents:
    displacy.render(sent, style='dep', jupyter=True)
```
